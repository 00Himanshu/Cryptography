 Write a Python function to check if a given integer is a prime number or not. Your function should take an integer as input and return ‘True’ if the number is prime, and ‘False’ otherwise. Additionally, provide an example demonstrating the usage of your function.





```python
def prime(n):
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

num = int(input("Enter a number: "))

if num <= 1:
    print(f"{num} is neither a prime nor a composite number")
else:
    if prime(num):
        print("It's a prime number")
    else:
        print("It is a composite number")
```

 Write a Python function that counts the number of prime factors of a given integer. Your function should take an integer as input and return the count of its prime factors.

```python
def factor(n):
    m = n
    factors_count = {}
    a = 2
    while True:
        if m == 1:
            break
        if m % a == 0:
            m //= a
            if a in factors_count:
                factors_count[a] += 1
            else:
                factors_count[a] = 1
        else:
            a += 1
    return factors_count

num=int(input("Enter a number: "))
print(factor(num))
```

Implement the Euclidean Algorithm in Python to find the greatest common divisor (gcd) of two given integers. Your function should take two integers as input and return their gcd.

```python
def gcd(n1, n2):
    if n1 > n2:
        dividend = n1
        divisor = n2
    else:
        dividend = n2
        divisor = n1
    while True:
        remainder = dividend % divisor
        if remainder == 0:
            return divisor
        else:
            dividend, divisor = divisor, remainder

a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))

gcd_result = gcd(a, b)
print(f"The gcd of {a} and {b} is {gcd_result}")
```

Write a Python function to find the least common multiple (LCM) of two given integers. Your function should take two integers as input and return their LCM.

```python
def lcm(n, m):
    g = gcd(n, m)
    return (n * m) // g

a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))

lcm_result = lcm(a, b)

print(f"The LCM of {a} and {b} is {lcm_result}")
```

```python
def mod_addition(n, a, b):
  return (a+b)%n

def mod_subtraction(n, a, b):
  return (a-b)%n

def mod_multiplication(n, a, b):
  return (a*b)%n
def mod_exponentiation(n, a, b):
  return (a**b)%n

n=int(input("Enter the value of n: "))
a=int(input("Enter the value of a: "))
b=int(input("Enter the value of b: "))

if n<=1:
  print("(n) should be a positive integer")
else:
  print(f"Modulo addition is: {mod_addition(n,a,b)}")
  print(f"Modulo subtraction is: {mod_subtraction(n,a,b)}")
  print(f"Modulo multiplication is: {mod_multiplication(n,a,b)}")
  print(f"Modulo exponentiation is: {mod_exponentiation(n, a, b)}")
```

2. Euler’s Totient Function ϕ(n):
     Write a Python function to find the number of invertible elements which are less than let's say n, and relatively prime
     to n  using Euler’s Totient function. Where n is a positive integer.

```python
def toient(n):
  result=[]
  for i in range(1,n):
    if (gcd(i,n))==1:
      result.append(i)
  return result

a=int(input("Enter number: "))
if a<=1:
  print("Number should be a positive integer")
else:
  result=toient(a)
  print(f"ϕ({a}) = {len(result)}")
```

```python
def Prime_test(n):
  prime=toient(n)
  if len(prime)==0:
    return False
  else:
    for i in prime:
      test=(2**(n-1))%n
      if test!=1:
        return False
  return True

num=int(input("Enter the number: "))

if num<=1:
  print(f"{num} is neither prime nor composite")

elif num == 2 or num == 3:
  print(f"{num} is prime")

elif Prime_test(num):
  print(f"{num} is prime")

else:
  print(f"{num} is composite")
```
