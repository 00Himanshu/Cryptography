```python
import binascii
from pyDes import des, CBC, PAD_PKCS5

def encrypt(plain_text, key):
    des_obj = des(key, CBC, key, pad=None, padmode=PAD_PKCS5)
    cipher_text = des_obj.encrypt(plain_text)
    return cipher_text.hex()

def decrypt(cipher_text, key):
    des_obj = des(key, CBC, key, pad=None, padmode=PAD_PKCS5)
    plain_text = des_obj.decrypt(bytes.fromhex(cipher_text))
    return plain_text.decode()

pt = "123456ABCD132536"
key = binascii.unhexlify("AABB09182736CCDD")

print("Encryption")
cipher_text = encrypt(pt, key)
print("Cipher Text:", cipher_text)

print("Decryption")
plain_text = decrypt(cipher_text, key)
print("Plain Text:", plain_text)
```
