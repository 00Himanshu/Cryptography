Implement a substitution cipher in Python where your first name is shifted by a value determined from the single-digit sum of your roll number.

```python
def substitution_cipher(name, roll):

    sum_roll = sum(int(i) for i in str(roll))

    while sum_roll>9:
      sum_roll = sum(int(i) for i in str(sum_roll))

    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    encrypted_name = ''

    for i in name.lower():
        if i.isalpha():
            index = alphabet.find(i)

            new_index = (index + sum_roll) % 26
            encrypted_name += alphabet[new_index]
        else:

            encrypted_name += i

    return encrypted_name.capitalize()

name = input("Enter you name: ").lower()

roll_number = int(input("Enter you roll no.: "))

encrypted_name = substitution_cipher(name, roll_number)

print(f"Encrypted name: {encrypted_name}")
```

Implement a Python program to perform encryption and decryption using the Caesar cipher.

```python
def caesar_cipher_encrypt(plaintext, shift):
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    encrypted_text = ''
    for i in plaintext:
        if i in alphabet:
            index = (alphabet.index(i) + shift) % len(alphabet)
            encrypted_text += alphabet[index]
        else:
            encrypted_text += i
    return encrypted_text

def caesar_cipher_decrypt(ciphertext, shift):
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    decrypted_text = ''
    for i in ciphertext:
        if i in alphabet:
            index = (alphabet.index(i) - shift) % len(alphabet)
            decrypted_text += alphabet[index]
        else:
            decrypted_text += i
    return decrypted_text


text = input("Enter the text to encrypt: ").lower()
shift = int(input("Enter the shift value: "))
encrypted_text = caesar_cipher_encrypt(text, shift)
print("Encrypted text:", encrypted_text)

decrypted_text = caesar_cipher_decrypt(encrypted_text, shift)
print("Decrypted text:", decrypted_text)
```
